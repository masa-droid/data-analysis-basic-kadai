```python
import numpy as np
A = np.array([[0,1],[2,3],[4,5]])
B = np.array([[0,1,2,3],[4,5,6,7]])
multi = np.dot(A,B)
print(multi)
maxmun = np.max(multi)
print(maxmun)
```

```python
import matplotlib.pyplot as plt
import numpy as np
fig, axes = plt.subplots(2,3,tight_layout=True)
y0 = [1,100,-5,2]
axes[0,0].plot(y0)

x1 = np.linspace(0, 10, 100)
y1 = 2 * np.sin(2 * x1)
axes[0,1].plot(x1,y1, linewidth=3.0)

x2 = np.random .normal(20,5,2000)
axes[0,2].hist(x2, bins=50)

np.random.seed(1)
x3 = np.random.normal(30,10,100)
y3 = np.random.normal(100,5,100)
axes[1,0].scatter(x3,y3)
axes[1,0].set(ylim=(0, 150), yticks=np.arange(0,150,20))



plt.show()
```
