```python

import japanize_matplotlib
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("sample_pandas_6.csv")

def tax(x):
    return x * 1.10


tax_series = df['単価'].apply(tax)
tax_series.name = "単価(税込み)"
df_concat = pd.concat([df,tax_series],axis=1)

category_df = pd.read_csv('category.csv')
df_concat = pd.merge(df_concat,category_df[['商品番号','カテゴリー']], how='inner',on='商品番号')

df_concat['商品名'].value_counts().plot(kind='bar')

plt.show()
df_concat.groupby('商品名')['注文数'].max().plot(kind='bar')
plt.show()
df_concat.groupby('商品名')['注文数'].min().plot(kind='bar')
plt.show()
df_concat.groupby('商品名')['注文数'].sum().plot(kind='bar')
plt.show()
df_concat.groupby('商品名')['注文数'].describe()
```
